```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os

from sklearn.preprocessing import OrdinalEncoder ,OneHotEncoder
```

```python
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
```

```python
print(f"Train Shape : {train.shape}")
print(f"Test Shape : {test.shape}")
```

```python
# Concat Two DataFrames
df = pd.concat([train , test] , axis = 0)
df.head()
```

```python
df.drop('CustomerID' , axis = 1 , inplace=True)
```

```python
df_encoded = df.copy()

ohe = OneHotEncoder(
    drop='first',
    sparse_output=False
)

gender_encoded = ohe.fit_transform(df_encoded[['Gender']])

gender_df = pd.DataFrame(
    gender_encoded,
    columns=ohe.get_feature_names_out(['Gender']),
    index=df_encoded.index
)

df_encoded = pd.concat([df_encoded.drop('Gender', axis=1), gender_df], axis=1)
```

```python
ordinal_cols = ['Subscription Type', 'Contract Length']

oe = OrdinalEncoder(
    categories=[
        ['Basic', 'Standard', 'Premium'],        
        ['Monthly', 'Quarterly', 'Annual']       
    ]
)

df_encoded[ordinal_cols] = oe.fit_transform(df_encoded[ordinal_cols])
```

```python
from sklearn.model_selection import train_test_split 

X = df_encoded.drop('Churn' , axis= 1)
y = df_encoded['Churn']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42, stratify=y
)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
train_df = pd.concat([X_train, y_train], axis=1)
test_df = pd.concat([X_test, y_test], axis=1)
```

```python
train_df = train_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)
```

```python
import pandas as pd
import datetime
from sklearn.model_selection import train_test_split
import os

# 1. Lấy lại DataFrame gốc chưa bị mã hóa One-Hot (nối train và test từ file CSV trên cùng lại)
raw_df = pd.concat([train, test], axis=0).dropna().drop_duplicates(subset=['CustomerID'], keep='first').copy()

# 2. Đổi lại tên cột cho khớp CHUẨN 100% với MLflow và Feast Config
rename_map = {
    'CustomerID': 'customer_id',
    'Age': 'age',
    'Gender': 'gender',
    'Tenure': 'tenure_months',
    'Usage Frequency': 'usage_frequency',
    'Support Calls': 'support_calls',
    'Payment Delay': 'payment_delay_days',
    'Subscription Type': 'subscription_type',
    'Contract Length': 'contract_length',
    'Total Spend': 'total_spend',
    'Last Interaction': 'last_interaction_days',
    'Churn': 'churned'
}
raw_df.rename(columns=rename_map, inplace=True)

# 3. Thêm timestamp bắt buộc cho Feast
current_time = pd.to_datetime(datetime.datetime.now())
raw_df['event_timestamp'] = current_time
raw_df['created_timestamp'] = current_time

# 4. Split dữ liệu lại với dữ liệu chữ nguyên thủy và tên chuẩn
X_raw = raw_df.drop('churned', axis=1)
y_raw = raw_df['churned']
X_train_raw, X_test_raw, y_train_raw, y_test_raw = train_test_split(
    X_raw, y_raw, test_size=0.15, random_state=42, stratify=y_raw
)

train_df_final = pd.concat([X_train_raw, y_train_raw], axis=1)
test_df_final = pd.concat([X_test_raw, y_test_raw], axis=1)

# 5. Lưu lại toàn bộ CSV và Parquet
train_df_final.to_csv("train.csv", index=False)
test_df_final.to_csv("test.csv", index=False)

os.makedirs("../../../data-pipeline/churn_feature_store/churn_features/feature_repo/data", exist_ok=True)
train_df_final.to_parquet("../../../data-pipeline/churn_feature_store/churn_features/feature_repo/data/train.parquet", index=False)
test_df_final.to_parquet("../../../data-pipeline/churn_feature_store/churn_features/feature_repo/data/test.parquet", index=False)
raw_df.to_parquet("../../../data-pipeline/churn_feature_store/churn_features/feature_repo/data/processed_churn_data.parquet", index=False)

print("Đã làm đẹp dữ liệu thành công! Hãy quay lại Terminal chạy .\run_sh\train.ps1 nhé!")
```
